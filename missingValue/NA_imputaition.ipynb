```python
import numpy as np 
import pandas as pd

#load data
train = pd.read_csv("../input/data-for-final-presentation/train.csv")
test = pd.read_csv("../input/final-revised/test1.csv")
train.head()
```

```python
train.drop(axis = 1,columns = "Unnamed: 0",inplace = True)
```

```python
test.drop(axis = 1 ,columns = 'X1',inplace = True)
```

```python
train.head()
```

```python
#eliminate category data

# Get the categorical and numeric columns
category_columns = [
    'ProductCD', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6',
                   'addr1', 'addr2', 'P_emaildomain', 'R_emaildomain', 'DeviceType', 'DeviceInfo',
                   'M1','M2','M3','M4','M5','M6','M7','M8','M9',"id_12",
                   "id_13","id_14","id_15","id_16","id_17","id_19","id_20",
                   "id_28","id_29","id_30","id_31","id_32","id_33",
                   "id_34","id_35","id_36","id_37","id_38"]

numeric_columns = list(set(train.columns) - set(category_columns))

#Dealing with those NAN value in numeric columns
check_if_null = train[numeric_columns].isnull().any()
train_null_num_cols = check_if_null[check_if_null].index


```

```python
na_values = {}
for n in train_null_num_cols:
    train[f'{n}_isna'] = train[n].isnull()
    median = train[n].median()
    train[n].fillna(median, inplace=True)
    na_values[n] = median

```

All NaN values have been removed. Now that's take a look with the memory

```python
train.memory_usage().sum()
```

 Downcast the columns to the lowest precision.

```python
int_columns = []
for c in numeric_columns:
    try:
        if train[c].fillna(-1.0).apply(float.is_integer).all():
            int_columns += [c]
    except Exception as e:
        print("error: ", c, e)
```

```python
stats = train[int_columns].describe().transpose()
stats
```

```python
train.dtypes
```

Some of them are float64, but some columns can be int8 or int16
So,let's downcast

```python
int8_columns = stats[stats['max'] < 256].index
int16_columns = stats[(stats['max'] >= 256) & (stats['max'] <= 32767)].index
```

```python
for c in int8_columns:
    train[c] = train[c].astype('int8')
    
for c in int16_columns:
    train[c] = train[c].astype('int16')
```

```python
train.memory_usage().sum()
```

```python
na_values.items()
```

```python
for k, v in na_values.items():
    test[f'{k}_isna'] = test[k].isnull()
    test[k].fillna(v, inplace=True)
```

```python
test.head()
```

```python
test_num_cols = list(set(numeric_columns) - set(['isFraud']))
test_check_if_null = test[test_num_cols].isnull().any()
test_null_num_cols = test_check_if_null[test_check_if_null].index
```

```python
for n in test_null_num_cols:
    test[n].fillna(train[n].median(), inplace=True)  # use the training set's median!
```

```python
int_cols = []
for c in test_num_cols:
    try:
        if test[c].fillna(-1.0).apply(float.is_integer).all():
            int_cols += [c]
    except Exception as e:
        print("error: ", c, e)
stats = test[int_cols].describe().transpose()
int8columns = stats[stats['max'] < 256].index
int16columns = stats[(stats['max'] >= 256) & (stats['max'] <= 32767)].index
for c in int8columns:
    test[c] = test[c].astype('int8')
    
for c in int16columns:
    test[c] = test[c].astype('int16')
```

Category

```python
for c in category_columns:
    train[c] = train[c].fillna("missing")
    
for c in category_columns:   
    test[c] = test[c].fillna("missing")
```

```python
train.head()
```

```python
cats = {}
for c in category_columns:
    train[c] = train[c].astype("category")
    train[c].cat.add_categories('unknown', inplace=True)
    cats[c] = train[c].cat.categories
```

```python
for k, v in cats.items():
    test[k][~test[k].isin(v)] = 'unknown'
```

```python
from pandas.api.types import CategoricalDtype

for k, v in cats.items():
    new_dtype = CategoricalDtype(categories=v, ordered=True)
    test[k] = test[k].astype(new_dtype)
```

```python
for c in category_columns:
    train[c] = train[c].cat.codes
    test[c] = test[c].cat.codes
```

```python
train.to_csv("train.csv")
```

```python
test.to_csv("test.csv")
```
